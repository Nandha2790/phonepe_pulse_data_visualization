```python
import pandas as pd
import json
import os
from urllib.request import urlopen
import pymysql
from sqlalchemy import create_engine
```

```python
# function to get aggerecared transanction data

def agg_trans(folder_path_aggerecated_trans):
    agg_trans_col = ["state","year","quarter","name","count_transaction","amount"]
    agg_trans_column = {i:[] for i in agg_trans_col}

    with urlopen("https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson") as response:
        state_geo_json = json.load(response)
    state = [state_geo_json['features'][i]['properties']['ST_NM'] for i in range(len(state_geo_json['features']))]
    state.sort()

    states = [f for f in os.listdir(folder_path_aggerecated_trans)]
    k = []
    agg_trans = []
    for i in range(len(states)):
        k.append(rf'{folder_path_aggerecated_trans}\{states[i]}')
        y = os.listdir(k[0])
        for j in range(len(y)):
            q = rf'{k[i]}\{y[j]}'
            qua = os.listdir(q)
            for l in range(len(qua)):
                with open(rf'{q}\{qua[l]}') as agg:
                    agg_trans = json.load(agg)
                    for m in range (len(agg_trans['data']['transactionData'])):
                        agg_trans_column['state'].append(state[i])
                        agg_trans_column['year'].append(y[j])
                        agg_trans_column['quarter'].append('Q'+(qua[l].replace('.json','')))
                        agg_trans_column['name'].append(agg_trans['data']['transactionData'][m]['name'])
                        agg_trans_column['count_transaction'].append(agg_trans['data']['transactionData'][m]['paymentInstruments'][0]['count'])
                        agg_trans_column['amount'].append(agg_trans['data']['transactionData'][m]['paymentInstruments'][0]['amount'])
    
    return pd.DataFrame(agg_trans_column)
```

```python
# function to get aggerecated user data

def agg_user(folder_path):
    agg_user_col = ["state","year","quarter",'registeredUsers','appOpens']
    agg_user_column = {i:[] for i in agg_user_col}

    with urlopen("https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson") as response:
        state_geo_json = json.load(response)
    state = [state_geo_json['features'][i]['properties']['ST_NM'] for i in range(len(state_geo_json['features']))]
    state.sort()

    states = [f for f in os.listdir(folder_path)]
    k = []
    agg_user = []
    for i in range(len(states)):
        k.append(rf'{folder_path}\{states[i]}')
        y = os.listdir(k[0])
        for j in range(len(y)):
            q = rf'{k[i]}\{y[j]}'
            qua = os.listdir(q)
            for l in range(len(qua)):
                with open(rf'{q}\{qua[l]}') as agg_users:
                    agg_user = json.load(agg_users)
                    if agg_user['data']['aggregated'] is not None:                            
                        agg_user_column['state'].append(state[i])
                        agg_user_column['year'].append(y[j])
                        agg_user_column['quarter'].append('Q'+(qua[l].replace('.json','')))
                        agg_user_column['registeredUsers'].append(agg_user['data']['aggregated']['registeredUsers'])
                        agg_user_column['appOpens'].append(agg_user['data']['aggregated']['appOpens'])
    return pd.DataFrame(agg_user_column)
```

```python
# function to get map transaction data

def map_trans(folder_path_map_trans):
    map_trans_col = ["state","year","quarter","dist_name","count_map","amount_map"]
    map_trans_column = {i:[] for i in map_trans_col}

    with urlopen("https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson") as response:
        state_geo_json = json.load(response)
    state = [state_geo_json['features'][i]['properties']['ST_NM'] for i in range(len(state_geo_json['features']))]
    state.sort()

    states = [f for f in os.listdir(folder_path_map_trans)]
    k = []
    map_trans = []
    for i in range(len(states)):
        k.append(rf'{folder_path_map_trans}\{states[i]}')
        y = os.listdir(k[0])
        for j in range(len(y)):
            q = rf'{k[i]}\{y[j]}'
            qua = os.listdir(q)
            for l in range(len(qua)):
                with open(rf'{q}\{qua[l]}') as map_tran:
                    map_trans = json.load(map_tran)
                    for m in range (len(map_trans['data']['hoverDataList'])):
                        map_trans_column['state'].append(state[i])
                        map_trans_column['year'].append(y[j])
                        map_trans_column['quarter'].append('Q'+(qua[l].replace('.json','')))
                        map_trans_column['dist_name'].append(map_trans['data']['hoverDataList'][m]['name'])
                        map_trans_column['count_map'].append(map_trans['data']['hoverDataList'][m]['metric'][0]['count'])
                        map_trans_column['amount_map'].append(map_trans['data']['hoverDataList'][m]['metric'][0]['amount'])
    
    return pd.DataFrame(map_trans_column)
```

```python
# function to get map user data
def map_user(folder_path_map_user):
    map_user_col = ["state","year","quarter","dist_name","registered_users","appOpens"]
    map_user_column = {i:[] for i in map_user_col}
    states = [f for f in os.listdir(folder_path_map_user)]

    with urlopen("https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson") as response:
        state_geo_json = json.load(response)
    state = [state_geo_json['features'][i]['properties']['ST_NM'] for i in range(len(state_geo_json['features']))]
    state.sort()

    k = []
    map_user = []
    for i in range(len(states)):
        k.append(rf'{folder_path_map_user}\{states[i]}')
        y = os.listdir(k[0])
        for j in range(len(y)):
            q = rf'{k[i]}\{y[j]}'
            qua = os.listdir(q)
            for l in range(len(qua)):
                with open(rf'{q}\{qua[l]}') as map_u:
                    map_user = json.load(map_u)
                    for m in range (len(map_user['data']['hoverData'])):
                        map_user_column['state'].append(state[i])
                        map_user_column['year'].append(y[j])
                        map_user_column['quarter'].append('Q'+(qua[l].replace('.json','')))
                        district_names = list(map_user['data']['hoverData'])
                        map_user_column['dist_name'].append(district_names[m])
                        map_user_column['registered_users'].append(map_user['data']['hoverData'][district_names[m]]['registeredUsers'])
                        map_user_column['appOpens'].append(map_user['data']['hoverData'][district_names[m]]['appOpens'])

                        
    
    return pd.DataFrame(map_user_column)
```

```python
# function for storing the data into MySQL
def export_Mysql(df,table_name):
    myconnection = pymysql.connect(host='127.0.0.1', user='root', passwd="user_password", database="phonepe_pulse") # replace user_password

    engine = create_engine('mysql+pymysql://root:user_password@127.0.0.1/phonepe_pulse') # replace user_password

    df.to_sql(f"{table_name}", con=engine, if_exists='replace', index=False)
    myconnection.close()
```
